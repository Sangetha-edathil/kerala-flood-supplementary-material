```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the file
file_path = 'C:/Users/SANGETHA/Downloads/THESIS/nino34.txt'
df = pd.read_csv(file_path, delim_whitespace=True)

# Create datetime column
df["Date"] = pd.to_datetime(df["YR"].astype(str) + "-" + df["MON"].astype(str).str.zfill(2))
df = df.sort_values("Date")

# Plot
plt.figure(figsize=(12, 4))
plt.plot(df["Date"], df["ANOM"], label='Nio 3.4 Anomaly (C)', color='tomato')
plt.axhline(0, color='gray', linestyle='--')

# Highlight August 2018
highlight_date = pd.to_datetime("2018-08-01")
plt.axvline(highlight_date, color='blue', linestyle='--', linewidth=1.5, label='Aug 2018')

plt.title("Monthly Nio 3.4 SST Anomaly (ERSSTv5)")
plt.xlabel("Year")
plt.ylabel("SST Anomaly (C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
